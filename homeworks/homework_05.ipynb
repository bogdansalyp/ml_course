### Standardization

```python
df['Fare'].mean()
```

```python
df['Fare'].std()
```

```python
fare_std = (df['Fare'] - df['Fare'].sum() / df['Fare'].shape[0]) / df['Fare'].std()
```

```python
fare_std.mean()
```

```python
fare_std.std()
```

```python
df['Fare'] = (df['Fare'] - df['Fare'].mean()) / df['Fare'].std()
```

```python
df['Age'] = (df['Age'] - df['Age'].mean()) / df['Age'].std()
```

```python
df.head()
```

### Null check

```python
df.isnull().values.any()
```

```python
pd.isna(df['Age']).sum()
```

```python
df['Age'][df['Sex'] == 0].fillna(df['Age'][df['Sex'] == 0].mean())
```

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
```

```python
df.isnull().values.any()
```

### Train/test features/target split

```python
df_X = df.drop(['Survived'], axis=1)
df_Y = df['Survived']
```

```python
df.shape
```

```python
X_train = df_X[:700]
X_test = df_X[700:]
Y_train = df_Y[:700]
Y_test = df_Y[700:]
```

### Model

```python
from sklearn.linear_model import LogisticRegression

clf = LogisticRegression()
clf.fit(X_train, Y_train)
```

```python
from sklearn.metrics import accuracy_score

print("train accuracy: {}".format(accuracy_score(Y_train, clf.predict(X_train))))
print("test accuracy: {}".format(accuracy_score(Y_test, clf.predict(X_test))))
```
