## Import libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

## Import the data

```python
pd.read_csv('../data/iris.csv')
```

```python
df = pd.read_csv('../data/iris.csv')
```

```python
df
```

## Preprocess the data

```python
df_train = df[df['species'] != 'virginica'][['sepal_length', 'petal_length', 'species']]
```

```python
df_train
```

```python
df_train['species'] = df_train['species'].astype('category').cat.codes
```

## Plot the data

```python
    plt.plot(df_train[df_train['species'] == 0]['sepal_length'], df_train[df_train['species'] == 0]['petal_length'], 'ro')
    plt.plot(df_train[df_train['species'] == 1]['sepal_length'], df_train[df_train['species'] == 1]['petal_length'], 'bo')
    plt.show()
```

## Create a model

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier()
```

```python
X_train = np.array(df_train[['sepal_length', 'petal_length']])
y_train = np.array(df_train['species'])
```

## Model learning

```python
knn.fit(X_train, y_train)
```

## Model predicting

```python
knn.predict([[6.5, 4]])
```

```python
h = 0.01
x_min, x_max = X_train[:, 0].min()-1, X_train[:, 0].max()+1
y_min, y_max = X_train[:, 1].min()-1, X_train[:, 1].max()+1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                     np.arange(y_min, y_max, h))
```

```python
y_train[0]
```

## Plot the decision boundary

```python
Z = knn.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
plt.plot(X_train[y_train == 0][:, 0], X_train[y_train == 0][:, 1], 'rx', label='Train - Setosa')
plt.plot(X_train[y_train == 1][:, 0], X_train[y_train == 1][:, 1], 'bx', label='Train - Versicolor')

plt.contourf(xx, yy, Z, cmap=plt.cm.RdBu, alpha=0.1) #  

plt.xlabel("Sepal length") #    X
plt.ylabel("Sepal width") #    Y
#plt.legend() #  
plt.show() #  
```
